```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"]= 1
df=dataset
```

```python
df[["Region"]] = df[["Region"]].astype(int)
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
#Remove. the white spaces from the column names

df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
# Generic form to get convert into other atatype

objects = [features for features in df.columns if df[features].dtypes == 'str']

for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.to_csv("Algerian_forest_fires_dataset_CLEANED.csv", index=False)
```

## Explotatory Data Analysis

```python
##Explotatory data analaysis

df_copy = df.drop(columns=["day", "month", "year"])
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'].value_counts()
```

```python
##Show some plots to understand the data better
import seaborn as sns
plt.style.use(sns.axes_style("whitegrid"))
df_copy.hist(bins=50,figsize=(20,15), color='blue')
plt.show()
```

```python
#3 Percentage for Pie chart


percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#Pie Chart

class_labels = ['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=class_labels, autopct='%1.1f%%', startangle=140, colors=['red', 'green'])
plt.title('Distribution of Fire and Not Fire Classes')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```
